# 使用分类数据绘图

- 分类变量在每个类别上显示的观测值
- 抽象出每个观测值的分布
- 应用统计估计值显示中心趋势测量值的观测值和置信区间

```python
%matplotlib inline

import matplotlib.pylab as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

## 分类散点图

```python
print(tips.iloc[:5])
sns.stripplot(x="day", y="total_bill", data=tips)
plt.show()
```

使用一些随机的“抖动”来调整位置(仅沿着分类轴)

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.show()
```

使用函数swarmplot(),它使用避免重叠点的算法定位分类上的每个散点1

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.show()
```

可以使用hue参数添加嵌套的分类变量

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex")
plt.show()
```

如果数据是pandas类型那么会默认按着pandas的既定顺序给出

```python
sns.swarmplot(x="size", y="total_bill", data=tips)
plt.show()
```

颠倒x和y的顺序来改变绘图方向

```python
sns.swarmplot(x="total_bill", y="day", hue="time", data=tips)
plt.show()
```

## 观测值分布

在多个类别之间进行快速比较的方法

### 箱型图

显示了分布的三个四分位数值及其极值，“晶须”延伸到位于下四分位数和上四分位数的1.5IQR内的点，落在该范围之外的数据独立显示

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
plt.show()
```

### 琴型图

```python
sns.violinplot(x="total_bill", y="day", hue="time", data=tips)
plt.show()
```

## 统计值估计

### 条形图

```python
sns.barplot(x="sex", y="survived", hue="class", data=titanic)
plt.show()
```

显示每个类别中观察值数量countplot()

```python
sns.countplot(x="deck", data=titanic, palette="Greens_d")
plt.show()
```

barplot()和countplot()一起调用

```python
sns.countplot(x="deck", hue="class", data=titanic, palette="Greens_d")
plt.show()
```

## 绘制多面板

使用factorplot()的优点是很容易绘图同时观测其他分类变量以此来对照

```python
sns.factorplot(x="time", y="total_bill", hue="smoker", col="day", 
               data=tips, kind="box", size=4, aspect=.5)
plt.show()
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker",
              col="time", data=tips, kind="swarm")
plt.show()
```
