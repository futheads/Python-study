# seaborn实例

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

## set_style()及set()

set_style( )是用来设置主题的，Seaborn有五个预设好的主题： darkgrid , whitegrid , dark , white ,和 ticks  默认： darkgrid

```python
sns.set_style('whitegrid')
plt.plot(np.arange(10))
plt.show()
```

set( )通过设置参数可以用来设置背景，调色板等，更加常用。

```python
sns.set(style='white', palette='muted', color_codes=True)
plt.plot(np.arange(10))
plt.show()
```

## distplot()与kdeplot()

distplot( )为hist加强版，kdeplot( )为密度曲线图 

```python
df_iris = pd.read_csv('./input/iris.csv')
print(df_iris.iloc[:5])
fig, axes = plt.subplots(1, 2)
sns.distplot(df_iris['sepal_length'], ax = axes[0], kde=True, rug=True)
sns.kdeplot(df_iris['sepal_length'], ax=axes[1], shade=True)
plt.show()
```

```python
sns.set(palette='muted', color_codes=True)
rs = np.random.RandomState(10)
d = rs.normal(size=100)
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot(d, kde=False, color='b', ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color='r', ax=axes[0, 1])
sns.distplot(d, hist=False, color='g', kde_kws={'shade': True}, ax=axes[1, 0])
sns.distplot(d, color='m', ax=axes[1, 1])
plt.show()
```

## 箱型图boxplot()

```python
df_iris = pd.read_csv('./input/iris.csv')
sns.boxplot(x=df_iris['species'], y=df_iris['sepal_length'])
plt.show()
```

```python
tips = pd.read_csv('./input/tips.csv')
sns.set(style='ticks')
print(tips.iloc[:5])
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='PRGn')
plt.show()
```

## 联合分布

```python
sns.jointplot('total_bill', 'tip', tips, kind='reg')
plt.show()
```

## 热点图heatmap()

```python
data = pd.read_csv('./input/tips.csv')
data = data.corr()
sns.heatmap(data)
plt.show()
```

## pairplot()

```python
data = pd.read_csv('./input/iris.csv')
sns.pairplot(data, hue='species')
plt.show()
```

```python
sns.pairplot(data, vars=['sepal_length', 'sepal_width'], hue='species', palette='husl')
plt.show()
```

## FacetGrid()

```python
tips = pd.read_csv('./input/tips.csv')
g = sns.FacetGrid(tips, col='time', row='smoker')
g = g.map(plt.hist, 'total_bill', color='r')
plt.show()
```
