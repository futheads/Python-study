```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import seaborn as sns
from pyecharts import Geo
import matplotlib.pyplot as plt
import jieba
```

```python
f = open("./input/狄仁杰_new.txt", encoding="utf-8")
data = pd.read_csv(f, sep=",", header=None, encoding="utf-8", names=["date", "nickname", "city", "rate", "comment"])
data[:5]
```

## 评分柱状图

```python
%matplotlib inline
```

```python
rate = data["rate"].value_counts()
sns.barplot(x=rate.index, y=(rate.values/sum(rate)))
```

```python
city = data.groupby(["city"])
rate_group = city["rate"]
city_com = city["city"].agg(["count"])
city_com.reset_index(inplace=True)
```

```python
# data_map = [(city_com["city"][i], city_com["count"][i]) for i in range(0, city_com.shape[0])]
data_map = [(city_com['city'][i],city_com['count'][i]) for i in range(0,city_com.shape[0])]
data_map[:5]
```

```python
geo = Geo("狄仁杰", title_color="#fff", title_pos="center", width=900, height=450, background_color="404a59")
```

```python
while True:
    try:
        attr, val = geo.cast(data_map)
        geo.add("", attr, val, visual_range=[0, 50], visual_text_color="#fff", is_geo_effect_show=False, is_piecewise=True, visual_split_number=6, symbol_size=15, is_visualmap=True)
    except ValueError as e:
        e = str(e)
        e = e.split("No coordinate is specified for ")[1]
        for i in range(0, len(data_map)):
            if e in data_map[i]:
                data_map.pop(i)
                break
    else:
        break
geo
```

```python
comment = jieba.cut(str(data["comment"]), cut_all=False)
wl_space_split = " ".join(comment)
background_image = plt.imread("./input/xuke.jpg")
stopwords = STOPWORDS.copy()
stopwords.add("电影")
stopwords.add("一部")
stopwords.add("一个")
stopwords.add("没有")
stopwords.add("什么")
stopwords.add("有点")
stopwords.add("这部")
stopwords.add("这个")
stopwords.add("不是")
stopwords.add("真的")
stopwords.add("感觉")
stopwords.add("觉得")
stopwords.add("还是")
stopwords.add("特别")
stopwords.add("非常")
stopwords.add("可以")
stopwords.add("因为")
stopwords.add("为了")
stopwords.add("比较")
```

```python
#设置词云参数
#参数分别是指定字体/背景颜色/最大的词的大小,使用给定图作为背景形状
wc = WordCloud(width=2048, height=2048, background_color='white',
              mask=background_image, font_path='C:/Windows/Fonts/simkai.ttf',
              stopwords=stopwords, max_font_size=400,
              random_state=50)
#将分词后数据传入云图
wc.generate_from_text(wl_space_split)
plt.imshow(wc)
plt.axis('off')#不显示坐标轴
```
