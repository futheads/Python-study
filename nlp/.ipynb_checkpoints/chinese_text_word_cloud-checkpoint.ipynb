```python
import jieba
import numpy
import codecs
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from wordcloud import WordCloud
```

```python
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (10, 5)
```

```python
df = pd.read_csv("./data/entertainment_news.csv", encoding="utf-8")
df.dropna()
content = df.content.values.tolist()

segment = []
for line in content:
    try:
        segs = jieba.lcut(line)
        for seg in segs:
            if len(seg) > 1 and seg != "\r\n":
                segment.append(seg)
    except:
        continue
```

```python
df_words = pd.DataFrame({"segment": segment})
stopwords = pd.read_csv("data/stopwords.txt", index_col=False, quoting=3, sep="\t", names=["stopword"], encoding="utf-8")
df_words = df_words[~df_words.segment.isin(stopwords.stopword)]
```

```python
stat_words = df_words.groupby(by="segment")["segment"].agg({"计数": numpy.size})
stat_words = stat_words.reset_index().sort_values(by=["计数"], ascending=False)
stat_words.head()
```

```python
wordcloud = WordCloud(font_path="data/simhei.ttf", background_color="white", max_font_size=80)
word_frequence = {x[0]: x[1] for x in stat_words.head(1000).values}
wordcloud = wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```
