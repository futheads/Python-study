## 查看数据的空间分布

```python
%matplotlib inline

from numpy import loadtxt, where
from pylab import scatter, show, legend, xlabel, ylabel

data = loadtxt("input/data1.txt", delimiter=",")

X = data[:, 0:2]
y = data[:, 2]

pos = where(y == 1)
neg = where(y == 0)

scatter(X[pos, 0], X[pos, 1], c="b")
scatter(X[neg, 0], X[neg, 1], c="r")

xlabel("Feature1/Exam 1 score")
ylabel("Feature2/Exam 2 score")

legend(["Fail", "Pass"])

show()
```
