# 快速开始

```python
from pyecharts import Bar

bar = Bar("我是第一个图表", "我是副标题")
bar.add("服装", ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"], [5, 20, 36, 10, 75, 90])
# bar.show_config()
bar
```

```python
bar = Bar("更多小工具", "这里是副标题")
bar.add("服装", ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"],\
        [5, 20, 36, 10, 75, 90], is_more_utils=True)
bar
```
